# Firstly merge all of these splitted csv files together

```python
import os
data_path = "../Dataset/BLE Data"
file_list = os.listdir(data_path)
```

```python
import pandas as pd
```

```python
#Load all the data with columns in the csv
csv_data = []
for csv_file in file_list:
  csv_data.append(pd.read_csv(os.path.join(data_path,csv_file),names=['user_id','timestamp','name', 'mac address', 'RSSI', 'power']))

#Concate all the data
df = csv_data[0]
for i in range(1,len(csv_data)):
  df = pd.concat([df,csv_data[i]])

print(df.shape)
df.head(5)
```

```python
# Save this fucking huge merged csv file for future use / no need to concatenate each time we 
# need to use ble data (15mins for the above cell of merging to execute)
df.to_csv("../cleaned_dataset/cleaned_ble_data.csv")
```
